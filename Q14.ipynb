```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist, squareform
from sklearn import metrics

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Compute the pairwise distance matrix
row_clusters = linkage(pdist(X, metric='euclidean'), method='complete')

# Create a dendrogram
plt.figure(figsize=(10, 6))
dendrogram(row_clusters, labels=iris.target, color_threshold=7)
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.title("Hierarchical Clustering Dendrogram (Iris Dataset)")
plt.axhline(y=7, color='gray', linestyle='--')  # Set a threshold for cutting the dendrogram
plt.show()

# Fit the Agglomerative Clustering model
n_clusters = 3
model = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='complete')
model.fit(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))
for i in range(n_clusters):
    plt.scatter(X[model.labels_ == i, 0], X[model.labels_ == i, 1], label=f"Cluster {i + 1}")

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Hierarchical Clustering of Iris Dataset")
plt.legend()
plt.show()
```
