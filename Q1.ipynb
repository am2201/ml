```python
import pandas as pd
import numpy as np
from sklearn import datasets
data = datasets.load_iris()
```

```python
X = data.data
y = data.target
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 1)
```

```python
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#x_train = sc.fit_transform(x_train)
#x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

cls = DecisionTreeClassifier(criterion = "entropy", random_state = 0)
cls.fit(x_train, y_train)
```

```python
fn= ['SepalLength(cm)', 'SepalWidth)(cm)', 'PetalLenght(cm)', 'PetalWidth(cm)']
cn = ['setosa', 'virginica', 'versicolor']
fig, axes = plt.subplots(nrows = 1, ncols =1, figsize = (6,6), dpi = 800)
tree.plot_tree(cls, feature_names=fn, class_names=cn, filled= True)
fig.savefig('DecisionTree_Iris.png')
```
